```python
import pandas as pd
```

```python
df=pd.read_csv("dft-road-casualty-statistics-casualty-provisional-mid-year-unvalidated-2022.csv")
```

```python
df.head()
```

```python
for column in df.columns:
    df[column]=df[column].replace([-1,"-1"],None)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df[df["pedestrian_location"]!=10]
df=df[df["pedestrian_movement"]!=9]
df=df[df["bus_or_coach_passenger"]!=9]
```

```python
X=df[["sex_of_casualty","age_band_of_casualty","pedestrian_location","pedestrian_movement","car_passenger","bus_or_coach_passenger"
      ,"casualty_home_area_type"]]
```

```python
Y=df["casualty_severity"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=4)
X_poly = poly_features.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_poly, Y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print('Mean Squared Error:', mse)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(Y_test, Y_pred)
print("R-squared:", r_squared)
```
