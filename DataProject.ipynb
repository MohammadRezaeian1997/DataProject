```python
import pandas as pd
```

```python
df=pd.read_csv("dft-road-casualty-statistics-casualty-provisional-mid-year-unvalidated-2022.csv")
```

```python
df.head()
```

```python
for column in df.columns:
    df[column]=df[column].replace([-1,"-1"],None)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df[df["pedestrian_location"]!=10]
df=df[df["pedestrian_movement"]!=9]
df=df[df["bus_or_coach_passenger"]!=9]
```

```python
numeric_columns = ['age_of_casualty']  # Assuming these are numeric columns

# Convert numeric columns to numeric data types
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)
```

```python
df.head()
```

```python
df['sex_of_casualty']
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
features = ['sex_of_casualty', 'age_band_of_casualty', 'pedestrian_location', 'pedestrian_movement',
            'car_passenger', 'bus_or_coach_passenger', 'casualty_home_area_type']
target = 'casualty_severity'
X = df[features]
y = df[target]
```

```python
label_encoder = LabelEncoder()
for column in X.select_dtypes(include=['object']):
    X[column] = label_encoder.fit_transform(X[column])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
importances = rf_classifier.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
```
